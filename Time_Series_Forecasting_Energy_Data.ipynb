## Train/Test split
Proper way is to use `cross-validation`

but we will split using random dates

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index >= '.01-01-2015']
# plot both data on subplots
fig, ax = plt.subplots(figsize=(15,5))
train.plot(ax=ax, label = 'Training Set')
test.plot(ax=ax, label='Test Set')
ax.axvline('01-01-2015', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
```

```python
df.loc[(df.index > '01-01-2010') & (df.index < '01-08-2010')].plot(figsize=(15,5), title='Week of Data')
```

## Feature Creation

```python
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.day_of_week
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df
```

```python
df = create_features(df)
```

## Visualize our Feature /Target Relationship

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='hour',y='PJME_MW')
ax.set_title('MW by Hour')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.boxplot(data=df, x='month',y='PJME_MW', palette='Blues')
ax.set_title('MW by Month')
plt.show()
```

## Create our Model

```python
# Train and Test features
train = create_features(train)
test = create_features(test)
```

```python
df.columns
```

```python
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year',
       'dayofyear']
TARGET = 'PJME_MW'
```

```python
# Create train and test datasets
X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

```python
# import metrics
# from sklearn.metrics import mean_squared_error
# ---
# We will be doing a regression test
reg = xgb.XGBRegressor(n_estimators=1000, early_stopping_rounds=50)
reg.fit(X_train, y_train, 
        eval_set=[(X_train,y_train), (X_test, y_test)], 
        verbose=True)
```

We can see for test set, the rms error decreases and starts to icrease, which is overfitting which we wanna avoid
so we can lower our learning rate

```python
reg = xgb.XGBRegressor(n_estimators=1000, early_stopping_rounds=50,
                      learning_rate=0.001)
reg.fit(X_train, y_train, 
        eval_set=[(X_train,y_train), (X_test, y_test)], 
        verbose=100)
# setting verbose to 100 will spit output at every 100 steps of learning
```

## Feature Importances

```python
reg.feature_importances_
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
            index=reg.feature_names_in_,
            columns=['importance'])
```

```python
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
```

## Forecast on Test set 
With the trained model

```python
test['prediction'] = reg.predict(X_test)
```

```python
test.head()
```

```python
df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)
```

```python
ax = df[['PJME_MW']].plot(figsize=(15,5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Prediction'])
ax.set_title('Raw Data and Predictions')
```

```python
ax = df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['PJME_MW'] \
.plot(figsize=(15,5), title='Week of Data')
df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['prediction'] \
.plot(style='.')
plt.legend(['Truth Data', 'Prediction'])
plt.show()
```

The predicted outputs are way out. We did not tune the parameters! But it's consistent with crest and valleys

```python
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f'RMSE Score on the test set: {score:0.2f}')
```

## Calculate Error

Lets look at the worst and best predicted days

```python
test['error'] = np.abs(test[TARGET] - test['prediction'])

# We want each date to have it's own value of error
test['date'] = test.index.date

# Get average error for each day that we predicted
test.groupby('date')['error'].mean().sort_values(ascending=False).head(5)
```

Setting `ascending=False` gives dta in descending order which is basically highest error first.
Doing the opposite, i.e `ascending=True` gives lowest error first

```python
test.groupby('date')['error'].mean().sort_values(ascending=True).head(5)
```
